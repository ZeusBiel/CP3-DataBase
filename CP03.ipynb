```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv("obitos-2024.csv")

# Verifique se há valores nulos nas variáveis
print(dados.isnull().sum())
```

```python
obitos_por_estado = dados.groupby('uf')['total'].sum()

# Criando gráfico de barras
obitos_por_estado.plot(kind='bar')
plt.xlabel('Estado')
plt.ylabel('Total de Óbitos')
plt.title('Óbitos por Estado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
obitos_por_estado = dados.groupby('faixa_etaria')['total'].sum()

# Criando gráfico de barras
obitos_por_estado.plot(kind='bar')
plt.xlabel('Faixa Etaria')
plt.ylabel('Total de Óbitos')
plt.title('Óbitos por Faixa Etária')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
eventos_por_tipo_doenca = dados['tipo_doenca'].value_counts().reset_index(name='Quantidade')
eventos_por_tipo_doenca.columns = ['Tipo Doença', 'Quantidade']

# Gráfico de barras para eventos por tipo de doença
plt.figure(figsize=(12, 6))
plt.bar(eventos_por_tipo_doenca['Tipo Doença'], eventos_por_tipo_doenca['Quantidade'])
plt.xlabel('Tipo de Doença')
plt.ylabel('Quantidade de Eventos')
plt.title('Distribuição de Eventos por Tipo de Doença')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
eventos_por_local_obito = dados['local_obito'].value_counts().reset_index(name='Quantidade')
eventos_por_local_obito.columns = ['Local Óbito', 'Quantidade']

# Gráfico de barras para eventos por local de óbito
plt.figure(figsize=(13, 7))
plt.bar(eventos_por_local_obito['Local Óbito'], eventos_por_local_obito['Quantidade'])
plt.xlabel('Local de Óbito')
plt.ylabel('Quantidade de Eventos')
plt.title('Distribuição de Eventos por Local de Óbito')
plt.tight_layout()
plt.show()
```

```python
# 5. Distribuição de Frequências por Sexo

# Contagem de eventos por sexo
eventos_por_sexo = dados['sexo'].value_counts().reset_index(name='Quantidade')
eventos_por_sexo.columns = ['Sexo', 'Quantidade']

# Gráfico de pizza para eventos por sexo
plt.figure(figsize=(8, 6))
plt.pie(eventos_por_sexo['Quantidade'], labels=eventos_por_sexo['Sexo'], autopct="%1.1f%%")
plt.title('Proporção de Eventos por Sexo')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```
